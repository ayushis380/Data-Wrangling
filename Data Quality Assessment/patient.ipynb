```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("patient.csv")
nan_ages = data['age_at_death'].isnull().sum()
nan_postal = data['postal_code'].isnull().sum()

to_plot = {'Columns': ['age_at_death', 'postal_code'],
        'Null Values': [nan_ages, nan_postal]
          }
df = pd.DataFrame(to_plot)

# Plotting the graph
ax = df.plot(x='Columns', kind='bar')
kx = -0.09
ky = 1
values =[nan_ages, nan_postal]
for i,rec in enumerate(ax.patches):
    ax.text(rec.get_xy()[0]+rec.get_width()/2+kx,rec.get_xy()[1]+rec.get_height()/2+ky,'{:.1%}'.format(values[i]/38), fontsize=12, color='black')

# Adding labels and titles
# plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.title('Percentage of Null Values in age_at_death and postal_code')
plt.legend(loc='lower left', fontsize =8)

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv("patient.csv")

each_val = data["race"].value_counts()
print(each_val)
plt.bar(each_val.index, each_val.values)
for i, v in enumerate(each_val.values):
    plt.text(i, v, str(round(v/38*100, 2))+"%", ha='center', va='bottom')
plt.xlabel("Different Types of Races", fontsize =14)
plt.ylabel("Total Count", fontsize =14)
plt.title("Percentage distribution for types of races")
plt.show()
```
